Pseudoco : scan intergers for largest 
Start
scan the the numbers intergers
12,4,27,3,188,50,7
Find the largest integers
output the the interger
End 

```python
nums = [12,4,27,3,758,50,7]

def find_largest (nums):
    if not nums:
        return "No value to compare"
    largest = nums[0]

    for n in nums[1:]:
        if n > largest:
            largest = n
    
    return largest
print(find_largest(nums))
   
```

Pseudoco : scan through intergers
Start
map through the list of integers to find all ods
the intergers are [3,4,10,15,999,2]
print the of intergers that are odd
End 

Assign1:
start
  Initualize counter to 0

  for each number in the list :
  if the number is an even interger:
  Increase counter by 1
  return counter 
end

```python
num_list =[3,5,27,20,6,3.5,60]

def num_total(num_list):
    total = 0
    for x in num_list:
        if x % 2 == 0:
            total += x
    return total
print(num_total(num_list))
```
