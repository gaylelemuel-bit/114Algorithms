```python
#C 1
class node:
    def __init__(self,value):
        self.value = value
        self.next = None 

#C 2
class LinkedList:
    def __init__(self):
        self.head = None        

    #C 3: append method
    def append(self, value):
        #add a new node at the end of the list
        new_node = node(value)
       # if the list is empty, new node becomes the head
        if self.head is None:
            self.head = new_node
            return    
        #otherwise, walk to the end and attach it
        current = self.head
        while current.next is not None:
            current = current.next

        current.next = new_node 

# C 3.1 Print List method 
    def print_list(self):
        current = self.head

        if current is None:
            print("List is empty!")
            return
        
        while current is not None:
            print(current.value)
            current = current.next 

#Create a linked list and append 5 value 
# #(10,15,20,15, an 30) 

    def count_nodes(self):
            count = 0
            current = self.head
            while current is not None:
                count += 1
                current = current.next
            return count 

    def prepend(self, value):
        new_node = node(value)
        new_node .next = self.head
        self.head = new_node 

# remove a specfic node 
    def remove(self,value):
        if self.head is None:
            return
        
        if self.head.value == value:
            self.head = self.head.next
            return
        
        current = self.head
        while current.next is not None:
            if current.next.value == value:
                current.next = current.next.next
                return
            current = current.next


my_list = LinkedList()


my_list.append(10)
my_list.append(15)
my_list.append(20)
my_list.append(25)
my_list.append(30)


print("Initial singly liked lise (after append):")
my_list.print_list()
print("node count:", my_list.count_nodes())
my_list.prepend(50)
my_list.print_list()
print("node count:", my_list.count_nodes())
my_list.remove(25)
my_list.print_list()
my_list.remove(12)



```

```python
# C: 1 node class
class node:
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None

# C: 2 doubly link list class
class DoublyLinkedList:
    def __init__(self):
        self.head = None

    #C: 3 append method
    def append(self, value):
        new_node = node(value)
       # check if the list is empty
        if self.head is None:
            # if it is empty, make new node
            self.head = new_node
            #stop
            return
        #Start traversing the list from head
        current = self.head

        #moving forward until we reach the last node
        while current.next is not None:
            current = current.next
        
        #link the last node to the new node
        current.next= new_node

        #link the new node back to the previous node
        new_node.prev = current 
    #C: 4
    def print_forward(self):
        current = self.head  # Start at the beginning of the list
        
        if current is None:
            print("The list is empty.")
            return
        
        while current:
            # Print the data; add an arrow if there's a next node
            print(current.value)
            current = current.next  # Move to the next node in the list
        
    #C: 5
    def print_backward(self):
        current = self.head  # Start at the beginning of the list
    
        if current is None:
            print("The list is empty.")
            return
        
        while current.next is not None:
            current = current.next
        while current:
        
            print(current.value)
            current = current.prev  
   

other_list = DoublyLinkedList()

other_list.append("A")
other_list.append("B")
other_list.append("C")
other_list.append("D")

print("Prints Forward:")
other_list.print_forward()
print("Prints Backward:")
other_list.print_backward()
```
